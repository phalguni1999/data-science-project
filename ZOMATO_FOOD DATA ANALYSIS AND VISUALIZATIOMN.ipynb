```python
#IMPORT ALL REQUIRED LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("dark_background")
```

```python
#READING THE DATASET(CSV FILE)

df = pd.read_csv("zomato.csv")
#to show the first five rows of the dataset
df.head()
```

```python
#TO SEE THE ROWS AND COLUMNS IN THE DATASET

df.shape
```

```python
df.size
```

```python
#TO SEE WHAT ARE THE COLUMNS PRESENT IN THED DATASET

df.columns
```

```python
#DROPPING THE COLUMNS WHICH ARE NOT IMPORTANT FOR THE ANALYSIS

df = df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis =1)

df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
#TO SEE THE INFORMATION ABOUT MY DATASET

df.info()
```

```python
#DROPPING THE DUPLICATED FROM THE DATASET

df.drop_duplicates(inplace = True)
df.shape
```

```python
#CLEANING THE RATE COLUMN IN THE DATASET

#(to see the unique values present in the rate column in the dataset)
df['rate'].unique()
```

```python
#NOW REMOVING THE UNWANTED THINGS FROM THIS COLUMN

def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
#TO CHECK THE NULL VALUES IN RATE COLUMN

df.rate.isnull().sum()
```

```python
#FILLING NULL VALUES IN RATE COLUMNWITH MEAN

df['rate'].fillna(df['rate'].mean() , inplace = True)
df.rate.isnull().sum()
```

```python
df.info()
```

```python
#DROPPING THE NULL VALUES

df.dropna(inplace = True)
df.head()
```

```python
#RENAMING THE LENGTHY COLUMNS NAME INTO SMALLER

df.rename(columns = {'approx_cost(for two people)':'cost2plates' , 'listed_in(type)':'type' , 'listed_in(city)':'city'} , inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['city'].unique()
```

```python
#BOTH 'LOCATION' AND 'CITY' COLUMNS HAVING SAME VALUE SO WE DON'T NEED BOTH THE COLUMN SO DROP ANY ONE COLUMN FROM THE DATASET

df = df.drop(['city'] , axis = 1)
df.head()
```

```python
df['cost2plates'].unique()
```

```python
#REMOVING COMMA FROM EACH VALUE WHICH ARE PRESENT IN THE COST2PLATES COLUMN

def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',' , '')
        return float(value)
    else:
        return float(value)
    
df['cost2plates'] = df['cost2plates'].apply(handlecomma)
df['cost2plates'].unique()
```

```python
df.head()
```

```python
#CLEANING REST_TYPE COLUMN IN THE DATASET

df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
#MAKING THE REST_TYPES LESS THAN 1000 IN FREQUENCIES AS OTHERS

def handle_rest_type(value):
    if (value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
#CLEANING THE LOCATION COLUMN IN THE DATASET

location = df['location'].value_counts(ascending = True)
```

```python
location_lessthan300 = location[location<300]

def handle_location(value):
    if (value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
#CLEANING THE CUISINES COLUMN

cuisines = df['cuisines'].value_counts(ascending = False)

cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if (value in cuisines_lessthan100):
        return 'others'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
